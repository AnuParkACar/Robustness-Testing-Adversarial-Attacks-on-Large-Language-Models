```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from datasets import load_dataset
from tqdm import tqdm
import evaluate
import random
```

# Model and Hardware Setup

```python
MODEL_NAME = "meta-llama/Llama-3.2-1B-Instruct"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
NUM_EXAMPLES = 10  # ***** Temporarily adjusted to 10 *****
MAX_NEW_TOKENS = 64
```

# Load LLM (Llama 3.2-1B)

```python
from transformers import pipeline

model_id = "meta-llama/Llama-3.2-1B-Instruct"

pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

# Load the Dataset (SQuAD)

```python
dataset = load_dataset("squad", split="validation[:1000]")
dataset = dataset.shuffle(seed=42)
dataset = dataset.select(range(NUM_EXAMPLES))
print(f"Loaded {len(dataset)} examples.")
```

# Define Adversarial Augmentation

```python
augmenter = CharSwapAugmenter(
    pct_words_to_swap=0.2,  # Swap 20% of words
    transformations_per_example=1
)
```

# Evaluate the Model

```python
import json

predictions = []
references = []

print("Evaluating...")
for item in tqdm(dataset):
    context = item["context"]
    question = item["question"]
    ground_truths = item["answers"]["text"]
    qid = item["id"]

    # Apply TextBugger adversarial modification to the question
    print(f"\nOriginal question:\t {question}")
    question = augmenter.augment(question)[0]
    print(f"Adversarial Question:\t {question}")

    prompt = [
        {"role": "system", "content": "Answer the question based on the following context: " + context},
        {"role": "user", "content": "Question: " + question},
    ]
    # print("\nprompt:\n", json.dumps(prompt, indent=4), "\n")

    outputs = pipe(
        prompt,
        max_new_tokens=MAX_NEW_TOKENS,
    )

    predicted_answer = outputs[0]["generated_text"][-1]["content"]
    # print(f"predicted_answer:\t {predicted_answer}")

    predictions.append({"id": qid, "prediction_text": predicted_answer})
    references.append({"id": qid, "answers": {"answer_start": [0], "text": ground_truths}})
```

# Results

```python
squad_metric = evaluate.load("squad")
results = squad_metric.compute(predictions=predictions, references=references)

print("\n\n--- Evaluation Results ---")
print(f"Exact Match (EM): {results['exact_match']:.2f}")
print(f"F1 Score: {results['f1']:.2f}")

print("\n\n--- Sample Predictions ---")
for pred, ref in zip(predictions[:10], references[:10]):
    print(f"Ground Truth(s): {ref['answers']['text']}")
    print(f"Predicted: {pred['prediction_text']}\n")
```
