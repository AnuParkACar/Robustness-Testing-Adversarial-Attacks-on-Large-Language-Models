```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from datasets import load_dataset
from tqdm import tqdm
import evaluate
import random
```

```python
!pip install textattack
from textattack.augmentation import CharSwapAugmenter
from textattack.augmentation.augmenter import Augmenter
from textattack.transformations import WordSwapWordNet
```

# Model and Hardware Setup

```python
MODEL_NAME = "mistralai/Mistral-7B-Instruct-v0.1"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
NUM_EXAMPLES = 1000  # ***** Temp Adjustment to 10 *****
MAX_NEW_TOKENS = 64
```

# Load and Quantize the LLM (Mistral-7B)

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)
```

# Define Helper Methods

```python
def format_prompt(context, question):
    return f"""<s>[INST]Answer the question based on the context below.

Context: {context}

Question: {question} [/INST]"""

def clean_answer(raw_response):
    """Extract the answer from the generation."""
    return raw_response.split("[/INST]")[-1].strip()
```

# Load the Dataset (SQuAD)

```python
dataset = load_dataset("squad", split="validation[:1000]")
dataset = dataset.shuffle(seed=42)
dataset = dataset.select(range(NUM_EXAMPLES))
print(f"Loaded {len(dataset)} examples.")
```

# Define Adversarial Augmentation

```python
adversarial_augmentation_enabled = True

augmenter = Augmenter(
    transformation=WordSwapWordNet(),
    pct_words_to_swap=0.2,
    transformations_per_example=1
)
```

# Evaluate the Model

```python
squad_metric = evaluate.load("squad")

predictions, references = [], []
DEBUG, first_item = False, True

print("Evaluating...")
for item in tqdm(dataset):
    context = item["context"]
    question = item["question"]
    ground_truths = item["answers"]["text"]
    qid = item["id"]

    if DEBUG or first_item:
      print(f"\nOriginal Question:\t {question}")

    # Apply WordSwapWordNet adversarial modification to the question
    if adversarial_augmentation_enabled:
      question = augmenter.augment(question)[0]
      if DEBUG or first_item:
        print(f"Adversarial Question:\t {question}")

    prompt = format_prompt(context, question)
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(DEVICE)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=MAX_NEW_TOKENS,
            pad_token_id=tokenizer.eos_token_id
        )

    raw_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
    predicted_answer = clean_answer(raw_output)

    predictions.append({"id": qid, "prediction_text": predicted_answer})
    references.append({"id": qid, "answers": {"answer_start": [0], "text": ground_truths}})

    first_item = False

results = squad_metric.compute(predictions=predictions, references=references)
```

# Dispaly Results

```python
print("\n--- Evaluation Results ---")
print(f"Exact Match (EM): {results['exact_match']:.2f}")
print(f"F1 Score: {results['f1']:.2f}")
```
